```python
from matplotlib import pyplot
from matplotlib.patches import Rectangle
from matplotlib.patches import Circle
from mtcnn.mtcnn import MTCNN
import time

def draw_boxes_on_image(filename, result):
  image = pyplot.imread(filename)
  pyplot.imshow(image)
  axes = pyplot.gca()
  
  for face in result:
    x, y, width, height = face['box']
    rectangle = Rectangle((x, y), width, height, fill=False, color='red')
    axes.add_patch(rectangle)

  for _, value in face['keypoints'].items():
    dots = Circle(value, radius=2, color='red')
    axes.add_patch(dots)

pyplot.show()

test_image = input("Insert Image: ")
print("Initializing...")
time.sleep(1)
pixels = pyplot.imread(test_image)

face_detector = MTCNN()
faces = face_detector.detect_faces(pixels)

draw_boxes_on_image(test_image, faces)
print(" ")
print("Measurements: ")
print(" ")
time.sleep(1)


faces
```
